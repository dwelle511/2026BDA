```python
#linspace(start,stop,num) @@
#일정한 간격의 값으로 배열 생성
#동일한 길이의 하위 간격으로 분할하는 수가 나열된 1차원 배열을 반환
#arrays with regularly-spaced
#전체 구간은 [시작, 끝]으로 기본적인 양쪽을 모두 포함, 일정한 간격을 구성하는 값인 배열의 수를 매개변수 num에 지정하여 기본값은 50
#np.linspace(start,stop,num) #num부분에 이제 step=2 즉 원소개수 2개라는거다
linsp=np.linspace(0,10,2)
linsp
```

```python
#start=포함, stop=포함, num=성성 점의 개수=step=(stop-start)/(num-1)\
stepNum=(10-0)/(2-1)
print("num구간의 개수아닌 크기",stepNum)

#np.arange= step크기 지정 vs np.linspace=점 개수 지정, 구간 정확힌 n등분시 유리함
```

```python
#linspace(start,stop,num)
#numpy.linspace(start,stop,num=50,endpoint=True,restep=False,dtype=None,axis=0)

#지정된 간격에 걸쳐 균등한 간격의 숫자를 반환
#endpoint: bool,optional, 끝점은 제외 될 수 있음=True, False=제외
# retstop=True: (samples,step)반환= step 샘풀 수 사이 간격= step크기, false=samples만 반환
```

## 3.6 브로드캐스팅

```python
#broadcasting: 산술 연산 ㅅ ,
a=np.array([3,4,5])
b=np.array([8,7,6])
a+b
```

```python
a=np.array([3,4,5])
b=np.array([8,7,6,6]) #error => operands could not be broadcast together with shapes (3,) (4,)
#두 배열 크기 달라 안됨
a+b
```

```python
#하나의 배열 이 1차원으로 다른 배열 열수 와 같은 구조의 브로드캐스팅

#(3,3)*(3,) or (1,3) 아래로 확장
#
from numpy import  arange
x=arange(1,10).reshape(3,3)
x
```

```python
y=arange(-1,2) #linspace 50=num
y
```

```python
x*y
```

```python
x+y
```

```python
#하나의 배열이 2차원으로 다른 배열 행수와 같은 구조의 브로드 캐스팅
#row-wise, normalization
from numpy import arange
x=arange(1,10).reshape(3,3)

y=arange(3,10,3).reshape(3,1)

np.set_printoptions(precision=2) #소수점 이하 자릿수를 2자리로 설정,2차원 배열

print(x)
print(y)
print(x/y)
print(x-y)
```

```python
#행 또는 열이 1인 두 배열의 브로드 캐스팅
```

```python
#상수 np.newaxis 활용한 배열 차원의 확장 =증가, 배열 x:(4,)인 1차원 배열
#x[None,:]또는 x[np.newaxis:], none 이 있는 축인 행이 추가, 모양 (1,4)의 2차원 배열의 보기 (view)
#가 반환, 즉 1차원에서 행이 추가디어 2차원으로 차원 증가
a=x[None,:]
a
```

```python
#연산 a+b수행, 배열 a는 열이 1이며 배열 b는 행이 1
a=x[:,np.newaxis]
a
```
